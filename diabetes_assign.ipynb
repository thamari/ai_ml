```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Reading data set to the python environment

data = pd.read_csv('diabetes.csv')
```

```python
data
```

```python
# Displaying the colums in the dataset

data.columns
```

```python
# Calculating the mean of each column in dataset

data.mean()
```

```python
# Check for the null values present in the dataset

data.isna().any()
```

```python
# Visualization using Histogram

plt.hist(data['Glucose'], bins=np.arange(0, 250, 50), color='red', rwidth=.95)
plt.title("Blood Glucose Level")
plt.show()
```

```python
# Visualization using Scatter

plt.scatter(data['Insulin'], data['Pregnancies'])
plt.title("Insulin Level VS Pregnancies")
plt.xlabel("INS")
plt.ylabel("PREG")
plt.show()
```

```python
# Visualization using Countplot

sns.countplot(x=data['DiabetesPedigreeFunction'])
plt.show()
```

```python
# Visualization using Pairplot

sns.pairplot(data[['Glucose', 'BloodPressure', 'Insulin', 'BMI', 'Age', 'Outcome']])
plt.show()
```
